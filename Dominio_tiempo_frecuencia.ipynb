```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import pandas as pd

# Parameters
fs = 100000  # Sampling frequency (increased to be > 2 * 3000 Hz)
T = 1/fs   # Sampling period
t = np.arange(0, 0.005, T) # Time vector

# Generate a sine wave
frequency = 3000  # Frequency of the sine wave (3 kHz)
amplitude = 1 # Amplitude of the sine wave (1V)
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the sine wave
plt.figure(figsize=(10, 4))
plt.plot(t, sine_wave)
plt.title('Seal Seno en el Dominio del Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Apply Fourier Transform
yf = fft(sine_wave)
xf = fftfreq(len(sine_wave), T)[:len(sine_wave)//2]

# Calculate amplitude in dB
amplitude_db = 20 * np.log10(np.abs(yf[0:len(sine_wave)//2]) / (len(sine_wave)/2) + 1e-9) # Add a small epsilon to avoid log(0)

# Find prominent frequencies (harmonics) and their amplitudes
# We can set a threshold to identify significant peaks
threshold_db = -50 # Example threshold in dB, adjust as needed
prominent_indices = np.where(amplitude_db > threshold_db)[0]
harmonic_frequencies = xf[prominent_indices]
harmonic_amplitudes_db = amplitude_db[prominent_indices]

# Create a table (Pandas DataFrame)
harmonics_table = pd.DataFrame({
    'Harmonic Frequency (Hz)': harmonic_frequencies,
    'Amplitude (dB)': harmonic_amplitudes_db
})

print("Tabla de Armnicos y Amplitudes (dB):")
display(harmonics_table)

# Plot the spectrum in dB
plt.figure(figsize=(10, 4))
plt.plot(xf, amplitude_db)
plt.title('Espectro de Frecuencia (dB)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud (dB)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.signal import square
import pandas as pd

# Parameters
fs = 100000  # Sampling frequency (increased to be > 2 * 3000 Hz)
T = 1/fs   # Sampling period
t = np.arange(0, 0.005, T) # Time vector

# Get duty cycle from user
while True:
    try:
        duty_cycle_percent = float(input("Ingresa el porcentaje del ciclo til del pulso (0-100): "))
        if 0 <= duty_cycle_percent <= 100:
            break
        else:
            print("Por favor, ingresa un valor entre 0 y 100.")
    except ValueError:
        print("Entrada invlida. Por favor, ingresa un nmero.")

duty_cycle = duty_cycle_percent / 100.0

# Generate a pulse wave (using square with duty cycle)
frequency = 3000  # Frequency of the pulse wave (3 kHz)
amplitude = 1 # Amplitude of the pulse wave (1V peak)
pulse_wave = amplitude * square(2 * np.pi * frequency * t, duty=duty_cycle)

# Plot the pulse wave
plt.figure(figsize=(10, 4))
plt.plot(t, pulse_wave)
plt.title(f'Seal de Pulso en el Dominio del Tiempo (Ciclo til: {duty_cycle_percent}%)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Apply Fourier Transform
yf = fft(pulse_wave)
xf = fftfreq(len(pulse_wave), T)[:len(pulse_wave)//2]

# Plot the spectrum in dB
amplitude_db = 20 * np.log10(np.abs(yf[0:len(pulse_wave)//2]) / (len(pulse_wave)/2) + 1e-9) # Add a small epsilon to avoid log(0)

# Find prominent frequencies (harmonics) and their amplitudes
# We can set a threshold to identify significant peaks
threshold_db = -50 # Example threshold in dB, adjust as needed
prominent_indices = np.where(amplitude_db > threshold_db)[0]
harmonic_frequencies = xf[prominent_indices]
harmonic_amplitudes_db = amplitude_db[prominent_indices]

# Create a table (Pandas DataFrame)
harmonics_table = pd.DataFrame({
    'Harmonic Frequency (Hz)': harmonic_frequencies,
    'Amplitude (dB)': harmonic_amplitudes_db
})

print("Tabla de Armnicos y Amplitudes (dB):")
display(harmonics_table)

plt.figure(figsize=(10, 4))
plt.plot(xf, amplitude_db)
plt.title('Espectro de Frecuencia (dB)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud (dB)')
plt.grid(True)
plt.show()
```
