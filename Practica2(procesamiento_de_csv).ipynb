```python

import pandas as pd
import matplotlib.pyplot as plt
import os
from google.colab import files
from IPython.display import display
import ipywidgets as widgets
import chardet
import re

# Configuración de matplotlib
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [14, 6]
plt.rcParams['axes.grid'] = True

# Función para detectar codificación
def detectar_codificacion(ruta):
    with open(ruta, 'rb') as f:
        return chardet.detect(f.read(10000))['encoding']

# Función mejorada para leer CSV
def leer_csv(ruta):
    encodings = ['utf-8', 'latin1', 'iso-8859-1', 'windows-1252']
    for enc in encodings:
        try:
            df = pd.read_csv(ruta, encoding=enc, on_bad_lines='skip')
            if not df.empty:
                return df
        except:
            continue
    return None

# Función para extraer el nombre base de la señal
def obtener_nombre_base(nombre_archivo):
    # Eliminar extensiones y sufijos como (1), (2), etc.
    nombre = os.path.splitext(nombre_archivo)[0]
    nombre = re.sub(r'\(\d+\)$', '', nombre)  # Eliminar (1), (2), etc.
    nombre = re.sub(r'-\s*(time|fft)', '', nombre, flags=re.IGNORECASE)
    return nombre.strip()

# Función para identificar columnas de señal
def identificar_columnas_señal(df):
    # Buscar columnas numéricas que no sean índices
    señales = []
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            if not df[col].is_monotonic_increasing:
                señales.append(col)

    # Si no encontramos, usar las últimas 2 columnas numéricas
    if not señales:
        num_cols = df.select_dtypes(include='number').columns
        señales = list(num_cols[-2:]) if len(num_cols) >= 2 else list(num_cols)

    return señales[:2]  # Tomar máximo 2 señales

# Función para graficar pares de señales
def graficar_par(time_df, fft_df, nombre_base):
    if time_df is not None and not time_df.empty:
        columnas_time = identificar_columnas_señal(time_df)
        plt.figure()
        for col in columnas_time:
            plt.plot(time_df.index, time_df[col], label=col)
        plt.xlabel('Tiempo')
        plt.ylabel('Amplitud')
        plt.title(f'Dominio de Tiempo: {nombre_base}')
        plt.legend()
        plt.grid(True)
        plt.show()

    if fft_df is not None and not fft_df.empty:
        columnas_fft = identificar_columnas_señal(fft_df)
        plt.figure()
        for col in columnas_fft:
            plt.plot(fft_df.index, fft_df[col], label=col)
        plt.xlabel('Frecuencia (Hz)')
        plt.ylabel('Magnitud')
        plt.title(f'Dominio de Frecuencia: {nombre_base}')
        plt.legend()
        plt.grid(True)
        plt.show()

# Procesamiento principal
def procesar_archivos(carpeta='/content'):
    archivos = sorted([f for f in os.listdir(carpeta) if f.lower().endswith('.csv')])

    if not archivos:
        print("No se encontraron archivos CSV.")
        return

    # Agrupar archivos por señal base
    archivos_por_señal = {}
    for archivo in archivos:
        nombre_base = obtener_nombre_base(archivo)
        if nombre_base not in archivos_por_señal:
            archivos_por_señal[nombre_base] = {'time': None, 'fft': None}

        if 'time' in archivo.lower():
            archivos_por_señal[nombre_base]['time'] = archivo
        elif 'fft' in archivo.lower():
            archivos_por_señal[nombre_base]['fft'] = archivo

    # Procesar cada grupo de señales
    for nombre_base, archivos in archivos_por_señal.items():
        print(f"\n{'='*50}")
        print(f"Procesando señal: {nombre_base}")

        time_df, fft_df = None, None

        # Cargar archivo TIME si existe
        if archivos['time']:
            ruta_time = os.path.join(carpeta, archivos['time'])
            time_df = leer_csv(ruta_time)
            if time_df is not None:
                print(f"\nArchivo TIME: {archivos['time']}")
                display(time_df.head(3))
            else:
                print(f"Error al leer {archivos['time']}")

        # Cargar archivo FFT si existe
        if archivos['fft']:
            ruta_fft = os.path.join(carpeta, archivos['fft'])
            fft_df = leer_csv(ruta_fft)
            if fft_df is not None:
                print(f"\nArchivo FFT: {archivos['fft']}")
                display(fft_df.head(3))
            else:
                print(f"Error al leer {archivos['fft']}")

        # Graficar si hay datos
        if time_df is not None or fft_df is not None:
            graficar_par(time_df, fft_df, nombre_base)
        else:
            print("No se encontraron datos válidos para graficar")

        print(f"{'='*50}")

# Ejecución principal
def main():
    print("PROCESADOR DE SEÑALES - TIME & FFT (Versión Mejorada)")
    print("Sube tus archivos CSV a /content o usa los existentes\n")

    # Opción para subir archivos
    files.upload()

    procesar_archivos()

if __name__ == "__main__":
    main()
```
